```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sample dataset: 3 inputs, corresponding to 2 outputs
X = np.array([[0, 0, 1],
              [1, 1, 0],
              [0, 1, 1],
              [1, 0, 0]])

Y = np.array([[1, 0],
              [0, 1],
              [1, 0],
              [0, 1]])

# Define the ANN model
model = Sequential([
    Dense(5, activation='relu', input_shape=(3,)),  # Hidden layer with 5 neurons
    Dense(2, activation='softmax')  # Output layer with 2 neurons
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, Y, epochs=100, verbose=1)

# Test prediction
print("Predictions:")
print(model.predict(X))

# b. Draw model
import matplotlib.pyplot as plt
import networkx as nx

def draw_ann():
    layers = [3, 5, 2]  # аё€аёіаё™аё§аё™а№‚аё«аё™аё”а№ѓаё™а№Ѓаё•а№€аёҐаё°аёЉаё±а№‰аё™ (Input, Hidden, Output)
    G = nx.DiGraph()  # аёЄаёЈа№‰аёІаё‡ Directed Graph

    pos = {}  # а№ЂаёЃа№‡аёљаё•аёіа№Ѓаё«аё™а№€аё‡а№‚аё«аё™аё”
    # аё§аёІаё”а№‚аё«аё™аё”а№ѓаё™а№Ѓаё•а№€аёҐаё°аёЉаё±а№‰аё™
    x_offset = 0
    for layer_idx, num_nodes in enumerate(layers):
        y_offset = (num_nodes - 1) / 2  # а№ѓаё«а№‰а№‚аё«аё™аё”аё­аёўаё№а№€аё•аёЈаё‡аёЃаёҐаёІаё‡
        for i in range(num_nodes):
            node_id = f"{layer_idx}-{i}"  # аё•аё±а№‰аё‡аёЉаё·а№€аё­а№‚аё«аё™аё” а№ЂаёЉа№€аё™ "0-0" (Input 1), "1-0" (Hidden 1)
            G.add_node(node_id, layer=layer_idx)
            pos[node_id] = (x_offset, y_offset - i)  # аёЃаёіаё«аё™аё”аё•аёіа№Ѓаё«аё™а№€аё‡а№‚аё«аё™аё”

        x_offset += 2  # а№Ђаё§а№‰аё™аёЈаё°аёўаё°аё«а№€аёІаё‡аёЈаё°аё«аё§а№€аёІаё‡аёЉаё±а№‰аё™

    # аё§аёІаё”а№ЂаёЄа№‰аё™а№ЂаёЉаё·а№€аё­аёЎаёЈаё°аё«аё§а№€аёІаё‡аёЉаё±а№‰аё™
    for layer_idx in range(len(layers) - 1):
        for i in range(layers[layer_idx]):  # аё™аёґаё§аёЈаё­аё™а№ѓаё™аёЉаё±а№‰аё™аё›аё±аё€аё€аёёаёљаё±аё™
            for j in range(layers[layer_idx + 1]):  # аё™аёґаё§аёЈаё­аё™а№ѓаё™аёЉаё±а№‰аё™аё–аё±аё”а№„аё›
                G.add_edge(f"{layer_idx}-{i}", f"{layer_idx + 1}-{j}")

    # аё§аёІаё”аёЃаёЈаёІаёџ
    plt.figure(figsize=(5, 5))
    nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray')
    plt.title("ANN-Multilayer Feedforward Network (1 hidden layer)")
    plt.show()

# а№ЂаёЈаёµаёўаёЃа№ѓаёЉа№‰аё‡аёІаё™аёџаё±аё‡аёЃа№ЊаёЉаё±аё™аё§аёІаё” ANN
draw_ann()
```
